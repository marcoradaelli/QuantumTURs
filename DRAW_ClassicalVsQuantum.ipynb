```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
file_in_classical = open("ClassicalData.dat")
rows_classical = file_in_classical.readlines()[1:]
v_t = []
v_Σ_L = []
v_Σ_R = []
v_I_LR = []

for row in rows_classical:
    t, Σ_L, Σ_R, I_LR = row.replace("\n","").split("\t")
    v_t.append(float(t))
    v_Σ_L.append(float(Σ_L))
    v_Σ_R.append(float(Σ_R))
    v_I_LR.append(float(I_LR))
    
v_t = np.array(v_t)
v_Σ_L = np.array(v_Σ_L)
v_Σ_R = np.array(v_Σ_R)
v_I_LR = np.array(v_I_LR)
```

```python
plt.plot(v_t, v_Σ_L * v_Σ_R, label="$\Sigma_L\Sigma_R$")
plt.plot(v_t, v_I_LR * v_I_LR, label="$I_{LR}$")
plt.plot(v_t, v_Σ_L * v_Σ_R - v_I_LR * v_I_LR, label="Difference")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("FI^2")
plt.title("Classical case")
plt.show()

v_difference_classical = v_Σ_L * v_Σ_R - v_I_LR * v_I_LR
```

```python
file_in_quantum = open("QuantumData.dat")
rows_quantum = file_in_quantum.readlines()[1:]
v_t = []
v_I_L = []
v_I_R = []
v_I_LR = []

for row in rows_quantum:
    t, I_L, I_R, I_LR = row.replace("\n","").split("\t")
    v_t.append(float(t))
    v_I_L.append(float(I_L))
    v_I_R.append(float(I_R))
    v_I_LR.append(float(I_LR))
    
v_t = np.array(v_t)
v_I_L = np.array(v_I_L)
v_I_R = np.array(v_I_R)
v_I_LR = np.array(v_I_LR)

plt.plot(v_t, v_I_L * v_I_R, label="$I_L I_R$")
plt.plot(v_t, v_I_LR * v_I_LR, label="$I_{LR}^2$")
plt.plot(v_t, v_I_L * v_I_R - v_I_LR * v_I_LR, label="Difference")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("FI^2")
plt.title("Quantum case")
plt.show()

v_difference_quantum = v_I_L * v_I_R - v_I_LR * v_I_LR
```

```python
plt.plot(v_t, v_difference_classical, label="Classical")
plt.plot(v_t, v_difference_quantum, label="Quantum")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("Bounds")
plt.savefig("ClassicalQuantumBounds.pdf")
plt.show()
```
