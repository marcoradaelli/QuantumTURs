```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
file_in_classical = open("ClassicalData_ThreeLevelsMaser.dat")
rows_classical = file_in_classical.readlines()[1:]
v_t = []
v_Σ_C = []
v_Σ_H = []
v_I_HC = []

for row in rows_classical:
    t, Σ_C, Σ_H, I_HC = row.replace("\n","").split("\t")
    v_t.append(float(t))
    v_Σ_C.append(float(Σ_C))
    v_Σ_H.append(float(Σ_H))
    v_I_HC.append(float(I_HC))
    
v_t = np.array(v_t)
v_Σ_C = np.array(v_Σ_C)
v_Σ_H = np.array(v_Σ_H)
v_I_HC = np.array(v_I_HC)
```

```python
plt.plot(v_t, v_Σ_C * v_Σ_H, label="$\Sigma_C\Sigma_H$")
plt.plot(v_t, v_I_HC * v_I_HC, label="$I_{HC}^2$")
plt.plot(v_t, v_Σ_C * v_Σ_H - v_I_HC * v_I_HC, label="Difference")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("FI^2")
plt.title("Classical case")
plt.show()

v_difference_classical = v_Σ_C * v_Σ_H - v_I_HC * v_I_HC
```

```python
file_in_quantum = open("QuantumData_ThreeLevelsMaser.dat")
rows_quantum = file_in_quantum.readlines()[1:]
v_t = []
v_I_C = []
v_I_H = []
v_I_HC = []

for row in rows_quantum:
    t, I_C, I_H, I_HC = row.replace("\n","").split("\t")
    v_t.append(float(t))
    v_I_C.append(float(I_C))
    v_I_H.append(float(I_H))
    v_I_HC.append(float(I_HC))
    
v_t = np.array(v_t)
v_I_C = np.array(v_I_C)
v_I_H = np.array(v_I_H)
v_I_HC = np.array(v_I_HC)
```

```python
plt.plot(v_t, v_I_C * v_I_H, label="$I_{CC} I_{HH}$")
plt.plot(v_t, v_I_HC * v_I_HC, label="$I_{HC}^2$")
plt.plot(v_t, v_I_C * v_I_H - v_I_HC * v_I_HC, label="Difference")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("FI^2")
plt.title("Quantum case")
plt.show()

v_difference_quantum = v_I_C * v_I_H - v_I_HC * v_I_HC
```

```python
plt.plot(v_t, v_difference_classical, label="Classical")
plt.plot(v_t, v_difference_quantum, label="Quantum")
plt.legend()
plt.xlabel("$t$")
plt.ylabel("Bounds")
plt.savefig("ClassicalQuantumBoundsThreeLevelsMaser.pdf")
plt.show()
```
